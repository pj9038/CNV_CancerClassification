```python
import pandas as pd
```

```python
tcga = pd.read_csv('common_tcga15.csv')
tcga.head()
```

```python
urine = pd.read_csv('Urine_data_k15.csv')
urine.head()
```

```python
urine=urine.dropna()
```

```python
urine=urine.loc[urine['y'].isin(['normal','bc'])]
tcga=tcga.replace(['BC', 'Normal'], ['bc', 'normal'])
tcga=tcga.loc[tcga['y'].isin(['normal','bc'])]
tcga.head()
```

```python
urine.drop(['cyto.15q11.1','cyto.1p36.33','cyto.21p11.1','cyto.22q13.33'],axis = 1,inplace=True)
```

```python
tcga=tcga[urine.columns]
```

```python
data=pd.concat([tcga, urine]) 
```

```python
X_train=data.iloc[0:10578,0:763]
Y_train=data.iloc[0:10578,763]
X_test=data.iloc[10578:10631,0:763]
Y_test=data.iloc[10578:10631,763]
```

```python
X_test.describe()
```

```python
X_train.describe()
```

# AdaBoost

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,MinMaxScaler,MaxAbsScaler
from sklearn.ensemble import AdaBoostClassifier
pipeline = Pipeline([('scaler', MaxAbsScaler()), ('classifier', AdaBoostClassifier())])
pipeline.fit(X_train, Y_train)
y_pred = pipeline.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy",(metrics.accuracy_score(Y_test, y_pred))*100,"%")
```

```python
Y_test #bc 31, normal 21
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(Y_test, y_pred)
```

```python
target_names = ['BC', 'Normal']
print(classification_report(Y_test, y_pred, target_names=target_names))
```
